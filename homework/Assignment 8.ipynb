```python
#The data from both files is read and saved to a variable using pandas.
ThermoelectricData = pd.read_csv("../dataset/CSC 593 dataset (thermoelectric).csv")
StrainData = pd.read_csv("../dataset/CSC 593 dataset (temp + strain).csv")

#Certain fields are derived from each dataset for the purposes of graphing. For last weeks homework I had
#trouble getting these fields but I figured out this week that my a field can be obtained using a notation
#similar to array indexing. Ex: file1['Field1'].
Time = ThermoelectricData['Time']
TFVoltage = ThermoelectricData['Thin film TC']
BVoltage = ThermoelectricData['Bulk TC']
HotT = ThermoelectricData['Hot junction (degC)']
ColdT = ThermoelectricData['Cold junction (degC)']
x = StrainData['Time']
y = StrainData['Resistance']

#Thermoelectric data versus time is plotted.
plt.figure()                                                    #A blank figure is generated.
plt.plot(Time,TFVoltage, 'r:', label = 'Thin film thermocouple')#First set of data is plotted on the figure.
plt.plot(Time,BVoltage, 'b--', label = 'Bulk thermocouple')     #Second set of data is plotted on the figure.
plt.xlabel('Time (s)')                                          #The x label name is set.
plt.ylabel('Voltage (V)')                                       #The y lable name is set.
plt.legend(loc = 'upper left')                                  #A legend is added to the top left of the plot.
plt.title('Thermoelectric voltages')                            #The plot title is set.
plt.grid()                                                      #A grid is added to the plot.

#Temperature data versus time is plotted.
plt.figure()                                                    #A blank figure is generated.
plt.plot(Time,HotT, 'r:', label = 'Hot junction')               #First set of data is plotted on the figure.
plt.plot(Time,ColdT, 'b--', label = 'Cold junction')            #Second set of data is plotted on the figure.
plt.xlabel('Time (s)')                                          #The x label name is set.
plt.ylabel('Temperature (degC)')                                #The y lable name is set.
plt.legend(loc = 'upper left')                                  #A legend is added to the top left of the plot.
plt.title("Junction temperatures")                              #The plot title is set.
plt.grid()                                                      #A grid is added to the plot.

#Resistance versus time is plotted.
plt.figure()                                                    #A blank figure is generated.
plt.plot(x, y, 'b')                                             #The data is plotted on the figure.
plt.xlabel('Time (s)')                                          #The x label name is set.
plt.ylabel('Resistance (ohm)')                                  #The y lable name is set.
plt.title('Resistance and time')                                #The plot title is set.
plt.grid()                                                      #A grid is added to the plot.
```

***Graph 1***: The first graph shows the thermoelectric response of two thermocouples as they are heated in a furnace over time. As the temperature increases the temperature dependent voltage also increases.

***Graph 2***: The second graph shows the temperature within the furnace and the temperature near the edge of the furnace as a function of time as the furnace program cycles.

***Graph 3***: The third graph shows the piezoresistive response of a semiconductor strain gage as it is mechanically deflected. Each oscillation corresponds the strain gage experiencing a mechanical force and then relaxing.
